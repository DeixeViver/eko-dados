# Educação Básica

### Cadastro das Escolas de Educaçăo Básica, com os níveis de ensino oferecidos

http://dados.gov.br/dataset/instituicoes-de-ensino-basico

```python
url = 'http://repositorio.dados.gov.br/educacao/CADASTRO_MATRICULAS_REGIAO_NORDESTE_PE_AL_SE_BA_2012.csv'

dados_url = requests.get(url).content
```

```python
edu_basica = pd.read_csv("dados/CADASTRO_MATRICULAS_REGIAO_NORDESTE_PE_AL_SE_BA_2012.csv", 
                  delimiter=';', header=11, encoding="ISO-8859-1", dtype='unicode')
edu_basica.head()
```

```python
edu_basica.columns.values
```

## Como se distribui a quantidade de escolas da rede pública e privada nos estados de Alagoas, Bahia, Pernambuco e Sergipe?

```python
edu_basica.groupby(['REDE', 'UF']).size().reset_index(name='Quantidade').sort_values('UF')
```

## E olhando a dependencia administrativa?

```python

edu_basica.groupby(['REDE', 'UF', 'Dependencia_Administrativa']).size().reset_index(name='Quantidade').sort_values('UF')
```

## E por localização?

```python
edu_basica.groupby(['REDE', 
                    'UF', 
                    'Dependencia_Administrativa', 
                    'Localizacao']).size().reset_index(name='Quantidade').sort_values('UF')
```
