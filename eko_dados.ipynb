## Visualizando respostas do formulário

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dados/dados_formulario.csv")
df.head()
```

```python
df.columns
```

### Vejamos alguns agrupamentos...

```python
campo = 'Carimbo de data/hora'
df.groupby(campo).size().reset_index(name='qtd').sort_values('qtd', ascending=False)
```

```python
campo = 'Qual a sua ideia ou como você vem atuando contra intolerância na Internet?'
df.groupby(campo).size().reset_index(name='qtd').sort_values('qtd', ascending=False)
```

```python
campo = 'Qual conhecimento ajudaria a fortalecer a sua ideia ou projeto?'
df.groupby(campo).size().reset_index(name='qtd').sort_values('qtd', ascending=False)
```

```python
campo = 'O que te impede de adquirir esse conhecimento com os recursos disponíveis na Internet hoje?'
df.groupby(campo).size().reset_index(name='qtd').sort_values('qtd', ascending=False)
```

```python
campo = 'Onde você costuma consumir esse tipo de conteúdo?'
df.groupby(campo).size().reset_index(name='qtd').sort_values('qtd', ascending=False)
```
